Firstly, we'd import the pandas and the pyplot for matplotlib. Then we'll read in <font color ='green'>Dataset 1: Smart Trash Containers</font> which displays the quantitiy of containers in each neighborhood. 

```python
import pandas as pd
from matplotlib import pyplot as plt
```

## DATASET 1: Smart Trash Containers ðŸ—‘

```python
smartTrashContainers = pd.read_csv('https://data.wprdc.org/datastore/dump/75b83ac9-8069-4cf1-bcc3-b9e6b04487d9')
zip_code = {}
for index, row in smartTrashContainers.iterrows():
    item = row['zip']
    if item in zip_code:
        zip_code[item] += 1
        continue
    else:
        zip_code[item] = 1
        
for key in zip_code:
    print(key)
    print(zip_code[key])
    print("--------")
```

**Conclusion**

From the data we can see that the top three zip codes with the most Smart Trash Containers are:
1. 15212 (Brighton Heights) with 166 trash containers
2. 15206 (Larimer) with 149 trash containers
3. 15219 (Downtown Pittsburgh) with 111 trash containers

## DATASET 3: Food Gardens ðŸŒ¾

Finally, we'll read in  <font color ='green'>Dataset 3: Food Gardens</font> and create a dictionary for it. Then we'd iterate through the dataframe, get the zip codes, amount of gardens in that zip code, and add to the qauantitiy if it's in the dictionary or set it to a quantity if it's not.

```python
foodGardens = pd.read_csv('https://data.wprdc.org/dataset/0d8c9708-3f01-48c4-adff-04d29825c9a2/resource/dbdfcb3e-8fa5-4468-9b05-f69562798f7a/download/growpghgardens201712.csv')

# Create a dictionary so for each zip code 
foodGardens_dict = dict()

# iterate through the dataframe
for i in range(len(foodGardens)):
    # get the zip code and quantity of gardens
    item = foodGardens.loc[i, 'zip_code']
    
    if item in foodGardens_dict: # if it's in the dict
        foodGardens_dict[item] += 1 # add the quantity to it
    else:
        foodGardens_dict[item] = 1 # otherwise, set it to the quantity
```

Then we'd convert the data into a series

```python
foodGardens_series = pd.Series(foodGardens_dict, name = "Gardens")
```

Finally, we'd sort and list the data in acending order

```python
foodGardens_series.sort_values()[:]
```

```python
# Created a series for the Food Gardens and removed Food Gardens not located in Pittsburgh, PA
greenways = pd.Series([1,1,2,4,2,3,3,4,4,6,5,5,7,13], index=["15223 (Sharpsburg)", "15203 (South Side)", "15217 (Squirrel Hill)", "15212 (Northside)", "15220 (Banksville)", "15224 (Garfield)", "15210 (Carrick)", "15207 (Hazelwood)", "15208 (Point Breeze)", "15213 (Oakland)", "15201 (Lawrenceville)", "15221 (Homewood)", "15219 (Schenley Heights)", "15206 (Larimer)"])

# Plot the bar graph
greenways.plot(kind="bar")
```

**Results:** Larimer ended up having the most Food Gardens with a total of 13. Following close behind were:
1. Schenley Heights w/ 7
